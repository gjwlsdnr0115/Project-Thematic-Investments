```python
def df_combine(df, lst):
    temp=[]
    for i in range(len(lst)):
        print(lst[i], end=' ')
        temp += df[df['Theme']==lst[i]]['Company'].iloc[0].replace(']','').replace('[','').replace(' ','').replace('\'', '').split(',')
        
        if i>=1:
            df.drop(index=df[df['Theme']==lst[i]].index, inplace=True)
    
    temp = list(set(temp))
    df.loc[df['Theme']==lst[0], 'Company'] = str(temp)
```
