# Extracting Theme List

```python
import pandas as pd
import numpy as np
import re
from tqdm import tqdm
```

```python
df = pd.read_csv("Data/news_data/final_data.csv", index_col=0)
df.head()
```

```python
kospi = pd.read_csv("Data/stock_data/kospi.csv", index_col=0, encoding='cp949')
kosdaq = pd.read_csv("Data/stock_data/kosdaq.csv", index_col=0, encoding='cp949')
```

```python
company_list = pd.concat([kospi[['종목코드', '기업명']], kosdaq[['종목코드', '기업명']]], axis=0)
company_list = company_list.sort_values('기업명').reset_index().drop(columns='번호')
company_list
```

```python
df.isnull().sum()
```

```python
df.drop(index=[1224], inplace=True)
df.reset_index(inplace=True)
```

```python
def theme_check(txt):
    return txt.endswith('한', '에', '는', '등', '며', '서')
```

```python
dic={}

for i in tqdm(range(len(df))):
    for company in company_list['기업명']:
        if company in df['contents'][i]:
            theme = 0
            try:
                theme_list = re.findall("[\s,\'](\w+?)\s[테마주,관련주]", df['title'][i])
                theme = re.findall("[\s,\'](\w+?)\s[테마주,관련주]", df['title'][i])[0]
                if theme.endswith(('한', '에', '는', '등', '며')):
                    theme = re.findall("[\s,\'](\w+?)\s[테마주,관련주]", df['contents'][i])[0]
                    if theme.endswith(('한', '에', '는', '등', '며')):
                        theme = re.findall("[\s,\'](\w+?)\s[테마주,관련주]", df['contents'][i])[1]
                try: dic[theme].append(company)
                except KeyError: dic[theme] = [company]
            except: pass
```

```python
# 중복 제거

cpy = dic

for key, value in dic.items():
    dic[key] = list(set(value))
```

```python
len(dic)
```

```python
theme_company_list = pd.DataFrame.from_dict(data=dic, orient='index')
theme_company_list
```

```python
theme_company_list.to_csv('data/theme_company_list.csv')
```
